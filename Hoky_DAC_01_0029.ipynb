# Data Preprocessing

## Data Train

### Missing Value

```python
datatrain = datatrain.dropna()
datatrain.reset_index(drop=True, inplace=True)
datatrain.index +=1
datatrain
```

### Duplicate Values

```python
duplicate_rows = datatrain[datatrain.duplicated()]
duplicate_rows
```

```python
#Drop Duplicates
datatrainclean = datatrain.drop_duplicates()
datatrainclean.reset_index(drop=True, inplace=True)
datatrainclean.index +=1
datatrainclean
```

# Feature Encoding


## Data Train

```python
datatrainclean
```

```python
datatrainclean.info()
```

```python
datatrain_object = datatrainclean.select_dtypes(include='object').columns
datatrain_object1 = datatrainclean.drop('type_of_attack', axis=1)
```

```python
label_encoder = LabelEncoder()

for kolom in datatrain_object1:
    datatrainclean[kolom] = label_encoder.fit_transform(datatrainclean[kolom])

datatrainclean.head()
```

## Data Test

```python
label_encoder = LabelEncoder()

for kolom_test in datatest_object:
    datatest[kolom_test] = label_encoder.fit_transform(datatest[kolom_test])

datatest.head()
```

# Splitting Data





```python
X = datatrainclean.drop('type_of_attack', axis=1)
Y =  datatrainclean['type_of_attack']
datatest = datatest.drop('id',axis=1)

X_train, X_test, y_train, y_test = train_test_split(X,Y, test_size=0.2, random_state=40)

X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

# Model Development

## SVM

```python
svm=SVC()
svm.fit(X_train,y_train)
```

```python
y_pred_svm = svm.predict(X_test)
y_pred_svm
```

## Model Predict

```python
result = svm.predict(datatest)
```

```python
data = datatest.copy()
data["type_of_attack"] = result
data.head()
```

## Accuracy

```python
accuracy_svm = svm.score(X_test,y_test)
print("Accuracy:", accuracy_svm)
```
